### Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

### Lets take 1000 sample and each sample of 50 observations.

```python
samples = np.random.exponential(scale=6.5,size=(1000,50))
```

### On each sample we have to find mvue value.

```python
mvue_values = np.sqrt(np.mean(samples**2,axis=1)/2)
```

```python
sns.distplot(a=mvue_values,bins=12,kde=True)

plt.xlabel("Mvue values of population parameter of  Exponential Distributed")

plt.ylabel("P(MVUE = k)")

plt.show()
```

### Checking unbiased estimate.

```python
unbiased_estimate = np.mean(mvue_values)
```
